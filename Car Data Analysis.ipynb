```python
import pandas as pd
```

### Fill the null value by mean val of the column for Cylinder col

```python
df['Cylinders'].fillna(df['Cylinders'].mean(), inplace = True)
```

```python
df.isnull().sum()
```

#### Show Car which are made in Asia or Europe

```python
df[df['Origin'].isin(['Asia', 'Europe'])]
```

### Delete all records where weight more than 4000

```python
df[~(df['Weight'] > 4000)]
```

### Increae the MPG_city value by 3 for all records

```python
df['MPG_City'] =  df['MPG_City'].apply(lambda x: x+3)
```
